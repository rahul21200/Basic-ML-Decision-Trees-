# Ensamble Techniques

## Bagging
Using the independent variables multiple number of times and making different trees from them. Thus, predicting the best possible decision from all of these trees combined

```python
from sklearn import tree
clftree = tree.DecisionTreeClassifier()
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bag_clf = BaggingClassifier(base_estimator=clftree , n_estimators=1000 , bootstrap=True , n_jobs=-1 , random_state=42)
```

```python
bag_clf.fit(X_train , Y_train)
```

```python
confusion_matrix(Y_test, bag_clf.predict(X_test))
```

```python
accuracy_score(Y_test, bag_clf.predict(X_test))
```

## Random Forests
Bagging may make similar trees each time as the number of strong affecting variables can be low and therefore we use subset of predictor variables such that we get different splits in each mode

If we take m variables to make a subset from p variables then:
1) for regression 
        m = P/3
2) for classification
        m = sqrt(P)

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_clf = RandomForestClassifier(n_estimators=1000, n_jobs=-1 , random_state=42)
```

```python
rf_clf.fit(X_train,Y_train)
```

```python
confusion_matrix(Y_test,rf_clf.predict(X_test))
```

```python
accuracy_score(Y_test,rf_clf.predict(X_test))
```

### Grid Search

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf_clf = RandomForestClassifier(n_estimators=250, random_state=42)
```

```python
params_grid = {"max_features" : [4,5,6,7,8,9,10] , "min_samples_split" : [2,3,10]}
```

```python
grid_search = GridSearchCV(rf_clf, params_grid, n_jobs=-1, cv = 5, scoring = "accuracy")
```

```python
grid_search.fit(X_train,Y_train)
```

```python
grid_search.best_params_
```

```python
cvrf_clf = grid_search.best_estimator_
```

```python
accuracy_score(Y_test , cvrf_clf.predict(X_test))
```

```python
confusion_matrix(Y_test , cvrf_clf.predict(X_test))
```
